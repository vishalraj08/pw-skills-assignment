Question 1: What is a random variable in probability theory? 

Answer:
A random variable is a function that assigns a numerical value to each outcome in the sample space of a random experiment.
 It's a variable whose possible values are numerical outcomes of a random phenomenon (like the result of a die roll)

Question 2: What are the types of random variables? 

Answer:

Discrete Random Variable: A variable that can take on a finite or a countably infinite number of values (e.g., number of defective products, always an integer).

Continuous Random Variable: A variable that can take on any value within a specified range or interval (e.g., height, temperature, time).

Answer:

Question 4: What is a binomial distribution, and how is it used in probability? 

Answer:
A binomial distribution is a discrete probability distribution that models the number of successes (k) in a fixed number of independent trials (n), where each trial has only two outcomes (success or failure) and the probability of success (p) is constant.
It is used to calculate the probability of a specific number of successes in scenarios that meet the BINS conditions: Binary outcomes, Independence, fixed Number of trials, and Same probability

Question 5: What is the standard normal distribution, and why is it important? 

Answer:
The standard normal distribution (Z-distribution) is a specific normal distribution with a mean (μ) of 0 and a standard deviation (σ) of 1.
It is important because any observation (x) from any normal distribution can be transformed into a Z-score using Z=(x−μ)/σ. This standardization allows statisticians to use a single table or function (the Z-table) to calculate probabilities and compare data from different datasets on a common scale

Question 7: What is the significance of confidence intervals in statistical analysis? 

Answer:
A confidence interval (CI) is a range of plausible values that is likely to contain the true population parameter (e.g., μ) with a specified level of confidence (e.g., 95%).
Its significance is that it:

Quantifies Uncertainty: It moves beyond a single point estimate (which is almost certainly wrong) to provide a measure of the precision of the estimate.

Aids Decision Making: It provides a range for the true parameter, which is more informative for business or scientific conclusions than a single value.

Question 8: What is the concept of expected value in a probability distribution? 

Answer:
The expected value (E[X]) is the long-run average value of a random variable. It is mathematically equivalent to the mean (μ) of the probability distribution. It represents the value you would 'expect' to obtain if the random experiment were repeated many times.

For a discrete variable, E[X]=∑x⋅P(X=x).

For a continuous variable, E[X]=∫x⋅f(x)dx.

Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean =50 and standard deviation =5 Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
%pip install matplotlib

import numpy as np

import matplotlib.pyplot as plt


# Parameters for the normal distribution
mu = 50       # Mean
sigma = 5     # Standard Deviation
n_samples = 1000  # Number of random numbers

# 1. Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=mu, scale=sigma, size=n_samples)

# 2. Compute its mean and standard deviation using NumPy
computed_mean = np.mean(data)
computed_std_dev = np.std(data)

# Output the results
output = (
    f"Generated Sample Size: {n_samples}\n"
    f"True Mean (mu): {mu}\n"
    f"True Standard Deviation (sigma): {sigma}\n"
    f"Computed Sample Mean: {computed_mean:.4f}\n"
    f"Computed Sample Standard Deviation: {computed_std_dev:.4f}\n"
)

# 3. Draw a histogram to visualize the distribution
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Sample Data')

# Add a line for the theoretical normal PDF for reference
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
# Normal PDF formula: P = 1 / (sigma * sqrt(2*pi)) * exp(-(x - mu)^2 / (2 * sigma^2))
p = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
plt.plot(x, p, 'r', linewidth=2, label='Theoretical PDF')

plt.title('Histogram of 1000 Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.axvline(computed_mean, color='green', linestyle='dashed', linewidth=1, label=f'Sample Mean: {computed_mean:.2f}')
plt.legend()
plt.grid(axis='y', alpha=0.4)
plt.show()

print(output)
```

Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]. Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval. Write the Python code to compute the mean sales and its confidence interval.

Application of the Central Limit Theorem (CLT)
The goal is to estimate the true average daily sales (μ) using the sample data. The Central Limit Theorem (CLT) is the underlying principle that allows us to do this using a sample of the data.


Selection of Distribution: Although the sample size is n=20 (less than the conventional 30), the CLT assures us that the distribution of sample means tends towards normality. Since the population standard deviation is unknown, we use the t-distribution for constructing the confidence interval, which is appropriate for small samples where the CLT still applies conceptually.

```python
import numpy as np
from scipy.stats import t

# Daily sales data (Sample) 
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# 1. Compute the sample statistics 
sample_mean = np.mean(daily_sales)
sample_std_dev = np.std(daily_sales, ddof=1) # ddof=1 for sample std dev
n = len(daily_sales)
confidence_level = 0.95
alpha = 1 - confidence_level

# 2. Calculate the t-critical value
degrees_of_freedom = n - 1
t_critical = t.ppf(1 - alpha/2, degrees_of_freedom)

# 3. Calculate the standard error and Margin of Error (ME)
standard_error = sample_std_dev / np.sqrt(n)
margin_of_error = t_critical * standard_error

# 4. Calculate the Confidence Interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Output the results
output = (\
    f"Sample Size (n): {n}\n"\
    f"Sample Mean (x̄): {sample_mean:.2f}\n"\
    f"Sample Standard Deviation (s): {sample_std_dev:.2f}\n"\
    f"Degrees of Freedom: {degrees_of_freedom}\n"\
    f"T-Critical Value (95% CI): {t_critical:.4f}\n"\
    f"Margin of Error: {margin_of_error:.2f}\n"\
    f"95% Confidence Interval for Mean Sales: ({ci_lower:.2f}, {ci_upper:.2f})\n"\
    f"Interpretation: We are 95% confident that the true average daily sales is between {ci_lower:.2f} and {ci_upper:.2f}."\
)

print(output)
```
